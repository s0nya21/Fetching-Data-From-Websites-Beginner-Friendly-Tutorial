```python
import requests #The requests module allows you to send HTTP requests using Python

from bs4 import BeautifulSoup

import pandas as pd
```

```python
# Get to know that which variables are essential for given case.

price = []
place = []
names = []
repair_info = []
bill_of_sale = []
mortgage = []

# Get data from any page you want by entering just relevant page number

page_num = input('Enter number')
for i in range(1, int(page_num)+1):
    url = 'https://bina.az/alqi-satqi/menziller?page=' + str(page_num)
    req = requests.get(url)
    content = BeautifulSoup(req.content, 'html.parser')  #taking in HTML code and extracting relevant information

    #The find_all method is used for finding out all tags with the specified tag name or id and returning them as a list of type bs4

    prices = content.find_all('span', {'class':'price-val'})
    location = content.find_all('div', {'class' : 'location'})
    name = content.find_all('ul', {'class' : 'name'})
    repair = content.find_all('div', {'class' : 'repair'})
    bill = content.find_all('div', {'class' : 'bill_of_sale'})
    mortg = content.find_all('div', {'class' : 'mortgage'})

    #inserting scraped data to lists for later use

    for i in mortg:
        mortgage.append(i.text)
    for i in bill:
        bill_of_sale.append(i.text)
    for i in repair:
        repair_info.append(i.text)
    for i in name:
        names.append(i.text)
    for i in prices:
        price.append(i.text)
    for i in location:
        place.append(i.text)
```

```python
df=pd.DataFrame() #creating dataframe with pandas
```

```python
len(place)
```

```python
#Defining columns

df['Names']=names
df['Place']=place
df['Repair_info']=pd.Series(repair_info)
df['Bill_of_sale']=pd.Series(bill_of_sale)
df['mortgage']=pd.Series(mortgage)
df['Price']=price
```
